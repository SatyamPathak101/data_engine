```python
df = df.drop('Name',axis=1)
df = df.drop('Date of Admission',axis=1)
df = df.drop('Doctor',axis=1)
df = df.drop('Hospital',axis=1)
df = df.drop('Insurance Provider',axis=1)
df = df.drop('Billing Amount',axis=1)
df = df.drop('Room Number',axis=1)
df = df.drop('Discharge Date',axis=1)
df = df.drop('Admission Type',axis=1)
```

```python
df.head(50)
```

```python
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(df)
```

```python
from sdv.single_table import GaussianCopulaSynthesizer

# Step 1: Create the synthesizer
synthesizer = GaussianCopulaSynthesizer(metadata)

# Step 2: Train the synthesizer
synthesizer.fit(df)

# Step 3: Generate synthetic data
synthetic_data = synthesizer.sample(num_rows=2000)
```

```python
from sdv.single_table import TVAESynthesizer

synthesizer = TVAESynthesizer(metadata)
synthesizer.fit(df)

synthetic_data = synthesizer.sample(num_rows=1500)
```

```python
synthesizer = TVAESynthesizer(
    metadata, # required
    enforce_min_max_values=True,
    enforce_rounding=False,
    epochs=1000
)
```

```python
synthesizer.get_parameters()
```

```python
metadata = synthesizer.get_metadata()
```

```python
synthesizer.fit(
    data=df
)
```

```python
synthesizer.get_loss_values()
```

```python
synthesizer.save(
    filepath='my_synthesizer.pkl'
)
```

```python
synthetic_data = synthesizer.sample(
    num_rows=10000,
    batch_size=10
)
```

```python
synthetic_data.to_csv('synthetic_data_tvae.csv', index=False)
```

```python
from sdv.evaluation.single_table import run_diagnostic, evaluate_quality
from sdv.evaluation.single_table import get_column_plot
import matplotlib.pyplot as plt

# 1. perform basic validity checks
diagnostic = run_diagnostic(df, synthetic_data, metadata)

# 2. measure the statistical similarity
quality_report = evaluate_quality(df, synthetic_data, metadata)

# 3. plot the data
fig = get_column_plot(
    real_data=df,
    synthetic_data=synthetic_data,
    metadata=metadata,
    column_name='Blood Type'
)
# fig.savefig('column_plot.png')
fig.show()
```

```python
categorical_features = ["Age", "Gender", "Blood Type", "Medical Condition", "Medication", "Test Results"]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your real dataset 'df' and synthetic dataset 'synthetic_data' already loaded
# Also assuming you have a list of categorical features 'categorical_features'

# Combine real and synthetic datasets with a 'Data Source' column
df['Data Source'] = 'Real Data'
synthetic_data['Data Source'] = 'Synthetic Data'
combined_data = pd.concat([df, synthetic_data])

# Plotting the distributions of categorical features
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='Data Source', data=combined_data)
    plt.title(f'{feature} Distribution: Real vs Synthetic')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Dataset')
    plt.show()
```
